```python
from getpass import getpass



# Class definitions


class FoodItem:
    # ... (same as before)
    def __init__(self, food_id, name, quantity, price, discount, stock):
        self.food_id = food_id
        self.name = name
        self.quantity = quantity
        self.price = price
        self.discount = discount
        self.stock = stock


class Admin:
    # ... (same as before)
    def __init__(self, username, password):
        self.username = username
        self.password = password

    def add_food_item(self, food_items, name, quantity, price, discount, stock):
        food_id = len(food_items) + 1
        food_item = FoodItem(food_id, name, quantity, price, discount, stock)
        food_items.append(food_item)
        print("Food Item added successfully!")

    def edit_food_item(self, food_items, food_id, name, quantity, price, discount, stock):
        found = False
        for food_item in food_items:
            if food_item.food_id == food_id:
                food_item.name = name
                food_item.quantity = quantity
                food_item.price = price
                food_item.discount = discount
                food_item.stock = stock
                print("Food Item edited successfully!")
                found = True
                break
        if not found:
            print("Food Item not found with the given FoodID.")

    def view_food_items(self, food_items):
        if len(food_items) > 0:
            print("List of Food Items:")
            for food_item in food_items:
                print(f"FoodID: {food_item.food_id}, Name: {food_item.name}, Quantity: {food_item.quantity},"
                      f" Price: {food_item.price}, Discount: {food_item.discount}, Stock: {food_item.stock}")
        else:
            print("No Food Items available.")

    def remove_food_item(self, food_items, food_id):
        found = False
        for food_item in food_items:
            if food_item.food_id == food_id:
                food_items.remove(food_item)
                print("Food Item removed successfully!")
                found = True
                break
        if not found:
            print("Food Item not found with the given FoodID.")


class User:
    # ... (same as before)
    def __init__(self, full_name, phone_number, email, address, password):
        self.full_name = full_name
        self.phone_number = phone_number
        self.email = email
        self.address = address
        self.password = password
        self.orders = []

    def place_order(self, food_items, selected_items):
        total_amount = 0
        for item in selected_items:
            food_item = food_items[item - 1]
            total_amount += (food_item.price - (food_item.price * food_item.discount / 100))
            food_item.stock -= 1
        self.orders.append(selected_items)
        print(f"Order placed successfully! Total Amount: INR {total_amount}")

    def view_order_history(self, food_items):
        if len(self.orders) > 0:
            print("Order History:")
            for order in self.orders:
                print("Items:")
                for item in order:
                    food_item = food_items[item - 1]
                    print(f"FoodID: {food_item.food_id}, Name: {food_item.name}, Quantity: {food_item.quantity},"
                          f" Price: {food_item.price}, Discount: {food_item.discount}, Stock: {food_item.stock}")
        else:
            print("No Order History available.")

    def update_profile(self, full_name, phone_number, email, address, password):
        self.full_name = full_name
        self.phone_number = phone_number

# Main function for running the app
def main():
    # Create some initial food items
    food_item1 = FoodItem(1, "Pizza", 1, 10.0, 0, 10)
    food_item2 = FoodItem(2, "Burger", 1, 5.0, 0, 20)
    food_item3 = FoodItem(3, "Fries", 1, 3.0, 0, 15)

    # Create an admin user
    admin = Admin("admin", "password")

    # Create a regular user
    user = User("John Doe", "1234567890", "john.doe@example.com", "1234 Elm Street", "password")

    while True:
        # Display main menu options
        print("\n===== Food Ordering App =====")
        print("1. Admin Login")
        print("2. User Login")
        print("3. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            # Admin Login
            print("\n===== Admin Login =====")
            username = input("Enter username: ")
            password = getpass("Enter password: ")
            if admin.username == username and admin.password == password:
                # Show admin menu
                while True:
                    print("\n===== Admin Menu =====")
                    print("1. Add Food Item")
                    print("2. Edit Food Item")
                    print("3. View Food Items")
                    print("4. Remove Food Item")
                    print("5. Logout")
                    admin_choice = input("Enter your choice: ")

                    if admin_choice == "1":
                        # Add Food Item
                        print("\n===== Add Food Item =====")
                        name = input("Enter food item name: ")
                        quantity = int(input("Enter quantity: "))
                        price = float(input("Enter price: "))
                        discount = float(input("Enter discount percentage: "))
                        stock = int(input("Enter stock: "))
                        admin.add_food_item(food_items, name, quantity, price, discount, stock)

                    elif admin_choice == "2":
                        # Edit Food Item
                        print("\n===== Edit Food Item =====")
                        food_id = int(input("Enter FoodID of the food item to edit: "))
                        name = input("Enter new food item name: ")
                        quantity = int(input("Enter new quantity: "))
                        price = float(input("Enter new price: "))
                        discount = float(input("Enter new discount percentage: "))
                        stock = int(input("Enter new stock: "))
                        admin.edit_food_item(food_items, food_id, name, quantity, price, discount, stock)

                    elif admin_choice == "3":
                        # View Food Items
                        print("\n===== View Food Items =====")
                        admin.view_food_items(food_items)

                    elif admin_choice == "4":
                        # Remove Food Item
                        print("\n===== Remove Food Item =====")
                        food_id = int(input("Enter FoodID of the food item to remove: "))
                        admin.remove_food_item(food_items, food_id)

                    elif admin_choice == "5":
                        # Logout
                        print("\nLogged out as Admin.")
                        break

                    else:
                        print("Invalid choice")
main()
```
